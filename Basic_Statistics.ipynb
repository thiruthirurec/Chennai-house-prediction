## Inferenctial Statistic

**HYPOTHESIS**

**Null hypothesis**: It is regarding the assumption that there is no anomaly pattern or believing according to the assumption made.

**Alternate hypothesis**: Contrary to the null hypothesis, it shows that observation is the result of real effect.



```python
from IPython import display
display.Image("./nullhypo.png", width="800", height="100")
```

**LEVEL OF SIGNIFICANCE:** 

**CONFIDENCE INTERVAL** 

Refers to the degree of significance in which we accept or reject the null-hypothesis. 100% accuracy is not possible for accepting or rejecting a hypothesis, so we therefore select a level of significance that is usually 5%. This is normally denoted with alpha(maths symbol ) and generally it is 0.05 or 5% , which means your output should be 95% confident to give similar kind of result in each sample.



**TYPES OF ERRORS**

**Type I error (  α ):** When we reject the null hypothesis, although that hypothesis was true. Type I error is denoted by alpha. In hypothesis testing, the normal curve that shows the critical region is called the alpha region.

**Type II errors (  β ):** When we failed to reject (* never say accept null hypo- stated in many papers) the null hypothesis but it is false. Type II errors are denoted by beta. In Hypothesis testing, the normal curve that shows the acceptance region is called the beta region.

**Paired T Test**

The paired sample t-test is also called dependent sample t-test. It’s an uni variate test that tests for a significant difference between 2 related variables. An example of this is if you where to collect the blood pressure for an individual before and after some treatment, condition, or time point.

H0 :- means difference between two sample is 0 (P value greater than 0.05)

H1:- mean difference between two sample is not 0 (P Value is smaller than 0.05)

Let us tests for a significant difference between weightof 2 Classes 'A' and 'B'

```python
import numpy as np
import scipy
from scipy import stats
```

```python
before_diet= np.array([72,65,83,45,48,90])
```

```python
After_diet= np.array([75,67,86,46,49,91])
```

So P value is lesser than 0.05 so accept H1 so there is effect in diet

```python
stats.ttest_rel(before_diet, After_diet)
```

** Dispersion of Data**

VARIANCE 

Variance is the numerical values that describe the variabilty of the observations from its arithmetic mean

Throwing 6 sided dice, D is Random variabe of Outcome of dice 

Expectation of D is E(D) = [1,2,3,4,5,6]

Maximum Variance = 1 - 6 = 5 

Minimum Variance = 1-2 = 1

Co- Variance 

Event A = E(A) = [3,3,3, 3]  ( 3 is a Mean / Variance is 0)

Event B = 
E(B) = [1,3,3,5] ( 3 is a Mean / Variance is 2)

Co-Variance is Positive, Negative and Neutral trend


```python
##1
import numpy as np
import statistics as stats
heights= [1,1,1,1,1,1,1,1,2] 

## Variance of dataset

Variance = np.var(heights)
Variance
```
